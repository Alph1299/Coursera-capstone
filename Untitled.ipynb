```python
import urllib, json
url = "https://services5.arcgis.com/GfwWNkhOj9bNBqoJ/arcgis/rest/services/NYC_Neighborhood_Tabulation_Areas/FeatureServer/0/query?where=1=1&outFields=*&outSR=4326&f=geojson"
nyc_data = gpd.read_file(url, driver='GeoJSON')
nyc_data['NTAName'] = nyc_data['NTAName'].astype('str')
```

```python
currentrent_data = pd.read_csv("data/medianAskingRent_All (2020).csv")
currentrent_data.head()
```

```python
merged_data = nyc_data.merge(currentrent_data, on='NTACode')
merged_data['NTAName'] = merged_data['NTAName'].astype('str')

merged_data.head()
```
