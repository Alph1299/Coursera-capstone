```python
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
x1 = [combined_df.Longitude]
y1 = [combined_df.Latitude]
```

```python
colors_map = np.array(['b', 'r'])
def assign_members(x1, y1, centers):
    compare_to_first_center = np.sqrt(np.square(np.array(x1) - centers[0][0]) + np.square(np.array(y1) - centers[0][1]))
    compare_to_second_center = np.sqrt(np.square(np.array(x1) - centers[1][0]) + np.square(np.array(y1) - centers[1][1]))
    class_of_points = compare_to_first_center > compare_to_second_center
    colors = colors_map[class_of_points + 1 - 1]
    return colors, class_of_points
```

```python
def update_centers(x1, y1, class_of_points):
    center1 = [np.mean(np.array(x1)[~class_of_points]), np.mean(np.array(y1)[~class_of_points])]
    center2 = [np.mean(np.array(x1)[class_of_points]), np.mean(np.array(y1)[class_of_points])]
    return [center1, center2]
```

```python
def plot_points(centroids=None, colors='g', figure_title=None):
    # plot the figure
    fig = plt.figure(figsize=(15, 10))  # create a figure object
    ax = fig.add_subplot(1, 1, 1)
    
    centroid_colors = ['bx', 'rx']
    if centroids:
        for (i, centroid) in enumerate(centroids):
            ax.plot(centroid[0], centroid[1], centroid_colors[i], markeredgewidth=5, markersize=20)
    plt.scatter(x1, y1, s=50, c=colors)
    
    # define the ticks
    xticks = np.linspace(42, 44, 44, endpoint=True)
    yticks = np.linspace(-80, -79, -79, endpoint=True)

    # fix the horizontal axis
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)

    # add tick labels
    xlabels = xticks
    ax.set_xticklabels(xlabels)
    ylabels = yticks
    ax.set_yticklabels(ylabels)

    # style the ticks
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.tick_params('both', length=2, width=1, which='major', labelsize=15)
    
    # add labels to axes
    ax.set_xlabel('Longitude', fontsize=20)
    ax.set_ylabel('Latitude', fontsize=20)
    
    # add title to figure
    ax.set_title(figure_title, fontsize=24)

    plt.show()
```

```python
plot_points(figure_title='Toronto Neighborhoods Geospatial')
```
